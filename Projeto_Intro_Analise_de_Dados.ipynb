**Desafio**

1. Empresa Vende Bermudas
2. 5 Lojas
3. Está querendo aumentar as vendas
4. O que fazer?
5. Informações Disponíveis: Base de Vendas

**Passo 1 - Trazer sua base de dados para o Python e ver o que tem nela**

```python
import pandas as pd
tabela = pd.read_excel('Vendas.xlsx')
display(tabela)
```

**Passo 2 - Pegar um panorama geral sobre a sua base de dados**

```python
faturamento_total = tabela["Valor Final"].sum()
print(faturamento_total)
```

**Passo 3 - Começar sua análise Top -> Down**

```python
# Faturamento por Loja
faturamento_loja = tabela[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
print(faturamento_loja)
```

**Passo 4 - Entrar no detalhe pra entender**

```python
faturamento_produto = tabela[["ID Loja","Produto", "Valor Final"]].groupby(["ID Loja","Produto"]).sum()
print(faturamento_produto)
```
